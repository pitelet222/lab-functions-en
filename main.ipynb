## 6. Write a function that returns the factorial of a number.

```python
def factorial(n):
    result = 1
    for i in range(1, n+1):
        result *= i
    return result
factorial(6)
```

```python
#factorial formula
#n! = n * ( n - 1 ) *...*1

# This code defines a function called "factorial" which takes an input "n". The function uses a for loop to iterate through the range of numbers 
# from 1 to n+1. For each number in that range, it multiplies the current value of x by the number in the range. At the end of the loop, 
# the function returns the final value of x, which will be the factorial of the input number "n".

# The Factorial of a positive integer n is the product of all positive integers less than or equal to n. 
# For example, the factorial of 6 (written "6!") is 6 * 5 * 4 * 3 * 2 * 1 = 720.

# So this function takes an input of any positive integer, and returns the factorial of that number.
```

## 7. Write a function that takes a list and returns a list of the unique values.

`NOTE: You cannot use set. ðŸ¤”`

```python
def unique(lst_un):
    uniqs = []
    for element in lst_un:
        found = False
        for u in uniqs:
            if u == element:
                found = True
                break
        if not found:
            uniqs.append(element)
    return uniqs

unique([1,2,3,2,3,2,3,2,3,4,5,3,5,3,2])
```

```python
# This will test your function 
test_unique(unique)
```

## 9. Write a function that calculates the standard deviation of a list.
`NOTE: Do not use any libraries or already built functions. ðŸ˜‰`

```python
def st_dev(list_sd):
    many = 0
    count = 0
    for n in list_sd:
        count += n
        many += 1
    mean = count / many

    difference_mean = 0
    for value in list_sd:
        difference_mean += (value - mean) ** 2
    variance = difference_mean / (many - 1)
    std_value = variance ** 0.5
    return std_value

st_dev([1,3,4,5,32,4,5,5,6,34,34,5,5,3,6,64,34,4,56,5,3])
    
```

```python
# This will test your function 
test_stdev(st_dev)
```

## 10. Write a function to check if a string is a pangram, i.e.: if it contains all the letters of the alphabet at least once. Mind that the strings may contain characters that are not letters.

```python
def pangram(string):
    abecedary = "abcdefghijklmnopqrstuvwxyz"
    abecedary_mayus = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    letters = []
   
     # iter through all the string and append the letters to a list called letters
    
    for letter in string:
        if letter in abecedary:
            letters.append(letter)
        elif letter in abecedary_mayus:
            idx = 0
            for i in abecedary_mayus:
                if i == letter:
                    break
                else:
                    idx += 1
            letters.append(abecedary[idx])          
     # Itera through all abecedary letters, if abecedary letter in the list letters, then true, else, false
    for abc in abecedary:
        encontrada = False
        for letra in letters:
            if letra == abc:
                encontrada = True
                break
    # if encontradas still false, return false
        if encontrada == False:
            return False
    return True
    
        
            
            
            
          
            
pangram("abcdefghijKlmnopqrstuvwxyZZZ")
```

## 12. Write a function to check if a given password is strong (at least 8 characters, at least one lower case, at least one upper case, at least one number and at least one special character). It should output True if strong and False if not.

```python
def check_pass(password):
    
    tiene_minus = False
    tiene_mayus = False
    tiene_special = False
    tiene_numero = False
    tiene_8 = False
        
    minus = "abcdefghijklmopqrstuvwxyz"
    mayus = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    numero = "1234567890"
    special = "!#$%&'()*+,-./:;<=>?@[]^_`{|}~"

    contador = 0
    
    for char in password:
        contador += 1
        if char in minus:
            tiene_minus = True
        if char in mayus:
            tiene_mayus = True
        if char in numero:
            tiene_numero = True
        if char in special:
            tiene_special = True
        if contador >= 8:
            tiene_8 = True
    if tiene_minus and tiene_mayus and tiene_numero and tiene_8 and tiene_special:
        return True
    else:
        return False

check_pass("fn}6&iIz")
        
```
